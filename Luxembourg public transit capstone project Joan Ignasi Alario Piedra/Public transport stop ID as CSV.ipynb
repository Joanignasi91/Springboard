<p><span style="font-size: 16pt;">Conversion public transport stop ID to CSV</span></p>
<p><span style="font-size: 11pt;">First I convert txt file to dataframe</span></p>

```python
import pandas as pd
df = pd.read_csv('Data\public transport stop ID\stop_ID.txt',sep=',', delimiter=';', header=None)
```

```python
# Read head of df to check for errors
df.head()
```

```python
# French notation is used for floats replace ',' for '.'
replace_df = pd.DataFrame(df[0].str.replace(',', '.'))
replace_df.head()
```

```python
# Datapoints are not split in columns, we now split the strings at '@'
split_series = replace_df[0].str.split('@')
split_series.head()
```

```python
# We convert the pandas series into a dictionary, no documentattion file could be found but appropriate headers can be deduced
# from the file, one can also deduce 'O' refers to name of stop or station, 'X' longitude east positive, 'Y' latitude north
# positive
frame_dict = {'id':[], 'O':[], 'X':[], 'Y':[], 'U':[], 'L':[], 'B':[], 'p':[]}
for i in split_series:
    frame_dict['id'].append(i[0])
    frame_dict['O'].append(i[1])
    frame_dict['X'].append(i[2])
    frame_dict['Y'].append(i[3])
    frame_dict['U'].append(i[4])
    frame_dict['L'].append(i[5])
    frame_dict['B'].append(i[6])
    frame_dict['p'].append(i[7])
basic_df = pd.DataFrame(frame_dict)
basic_df.head()
```

```python
# We keep this format for the moment, we save the data frame as a csv file, we check if saving went correctly
basic_df.to_csv('Data\public transport stop ID\df_stop_ID.csv', index=False, header=True)
pd.read_csv('Data\public transport stop ID\df_stop_ID.csv', nrows=5)
```
